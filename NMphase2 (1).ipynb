```python

cat_cols = df.select_dtypes(include='object').columns
if len(cat_cols) > 0:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=cat_cols[0])
    plt.title(f"Count Plot of {cat_cols[0]}")
    plt.xticks(rotation=45)
    plt.show()
```

```python

# Generate random data
import numpy as np # Import the numpy library with alias np
np.random.seed(10)
data = np.random.normal(100, 20, 200)

# Create box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Normally Distributed Data')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python

# Labels and sizes
labels = ['Segment A', 'Segment B', 'Segment C', 'Segment D']
sizes = [25, 35, 20, 20]

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Segments')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
